## 4. Boucles : repeter des operations

### La boucle `for`

La boucle `for` parcourt les elements d'une sequence (liste, range, chaine...).

```python
# Parcourir une liste
villes = ["Paris", "Lyon", "Marseille"]

for ville in villes:
    print(f"Visite de {ville}")
```

```python
# Parcourir avec range() : genere une suite d'entiers
for i in range(5):
    print(i, end=" ")

print()  # Retour a la ligne

# range(debut, fin, pas)
for i in range(0, 10, 2):
    print(i, end=" ")
```

```python
# enumerate() : obtenir l'indice ET la valeur
villes = ["Paris", "Lyon", "Marseille", "Nantes"]

for indice, ville in enumerate(villes):
    print(f"Ville {indice} : {ville}")
```

### La boucle `while`

La boucle `while` repete un bloc tant qu'une condition est vraie. Elle est utile quand on ne connait pas a l'avance le nombre d'iterations (par exemple, dans un algorithme de Dijkstra ou un backtracking).

```python
# Exemple : diviser par 2 jusqu'a atteindre 1
n = 64
compteur = 0

while n > 1:
    n = n // 2
    compteur += 1
    print(f"Etape {compteur} : n = {n}")

print(f"Il a fallu {compteur} divisions pour atteindre 1.")
```

### `break` et `continue`

- `break` : sort immediatement de la boucle
- `continue` : passe directement a l'iteration suivante

```python
# Trouver la premiere ville dont le nom depasse 5 caracteres
villes = ["Paris", "Lyon", "Marseille", "Lille", "Nantes"]

for ville in villes:
    if len(ville) > 5:
        print(f"Trouvee : {ville}")
        break

# Afficher uniquement les distances superieures a 400
distances = [220, 465, 385, 775, 150]

for d in distances:
    if d <= 400:
        continue
    print(f"Grande distance : {d} km")
```

### Boucles imbriquees

Les boucles imbriquees sont indispensables pour parcourir des paires d'elements (par exemple, toutes les paires de villes pour calculer une matrice de distances).

```python
villes = ["A", "B", "C"]

print("Toutes les paires de villes distinctes :")
for i in range(len(villes)):
    for j in range(i + 1, len(villes)):
        print(f"  {villes[i]} -- {villes[j]}")
```

## 5. Les tuples

Un tuple est similaire a une liste, mais **immutable** : une fois cree, on ne peut plus modifier son contenu. C'est ideal pour representer des donnees qui ne changent pas, comme les coordonnees d'une ville ou une arete d'un graphe.

```python
# Coordonnees d'une ville
coordonnees_paris = (48.8566, 2.3522)
print("Latitude :", coordonnees_paris[0])
print("Longitude :", coordonnees_paris[1])

# Deballage (unpacking) : extraire les valeurs dans des variables
latitude, longitude = coordonnees_paris
print(f"Paris est a la latitude {latitude} et longitude {longitude}")
```

```python
# Un tuple est immutable : on ne peut pas le modifier
try:
    coordonnees_paris[0] = 50.0
except TypeError as e:
    print(f"Erreur attendue : {e}")

# Cas d'usage : representer une arete ponderee (voisin, distance)
arete = ("Lyon", 465)
voisin, distance = arete
print(f"Voisin : {voisin}, Distance : {distance} km")
```

## 6. Les dictionnaires

Le dictionnaire associe des **cles** a des **valeurs**. C'est la structure de choix pour representer un graphe par liste d'adjacence : chaque cle est un sommet, et sa valeur est la liste de ses voisins.

```python
# Creer un dictionnaire : population de villes
populations = {
    "Paris": 2_161_000,
    "Lyon": 522_000,
    "Marseille": 873_000,
}

# Acceder a une valeur par sa cle
print("Population de Lyon :", populations["Lyon"])

# Ajouter une entree
populations["Nantes"] = 320_000

# Tester si une cle existe
print("Lille est dans le dictionnaire :", "Lille" in populations)
```

```python
# Parcourir un dictionnaire

# Les cles seules
for ville in populations:
    print(ville)

print("---")

# Les cles et les valeurs ensemble
for ville, pop in populations.items():
    print(f"{ville} : {pop:,} habitants")
```

```python
# Methode .get() : eviter une erreur si la cle n'existe pas
print(populations.get("Strasbourg", 0))  # Retourne 0 si absent

# Liste des cles et des valeurs
print("Cles :", list(populations.keys()))
print("Valeurs :", list(populations.values()))
```

### Application : representer un graphe

Un graphe pondere non oriente se represente naturellement par un dictionnaire ou chaque sommet est associe a la liste de ses voisins (sous forme de tuples `(voisin, poids)`).

```python
graphe = {
    "A": [("B", 10), ("C", 5)],
    "B": [("A", 10), ("C", 8)],
    "C": [("A", 5), ("B", 8)],
}

# Afficher les voisins de chaque sommet
for sommet, voisins in graphe.items():
    noms_voisins = [v for v, _ in voisins]
    print(f"Sommet {sommet} -> voisins : {noms_voisins}")
```

## 7. Les ensembles (set)

Un ensemble est une collection **sans doublons** et **sans ordre**. Il est ideal pour memoriser les sommets deja visites dans un parcours de graphe.

```python
# Creer un ensemble
sommets_visites = set()

# Ajouter des elements
sommets_visites.add("A")
sommets_visites.add("B")
sommets_visites.add("A")  # Pas de doublon : A est ignore

print("Visites :", sommets_visites)
print("Nombre de sommets visites :", len(sommets_visites))

# Test d'appartenance (tres rapide : O(1) en moyenne)
print("A visite ?", "A" in sommets_visites)
print("C visite ?", "C" in sommets_visites)
```

```python
# Operations ensemblistes
tous_les_sommets = {"A", "B", "C", "D"}
sommets_visites = {"A", "B"}

# Sommets restant a visiter
restants = tous_les_sommets - sommets_visites
print("Restants :", restants)

# Intersection
groupe_1 = {"A", "B", "C"}
groupe_2 = {"B", "C", "D"}
print("En commun :", groupe_1 & groupe_2)

# Union
print("Tous reunis :", groupe_1 | groupe_2)
```

### Representer une arete de maniere canonique

Dans un graphe non oriente, l'arete A--B est la meme que B--A. Pour eviter de la compter deux fois, on peut la stocker sous forme de tuple trie dans un ensemble.

```python
aretes_vues = set()

# Ajouter l'arete (C, A) sous forme canonique
arete = tuple(sorted(("C", "A")))
aretes_vues.add(arete)
print("Arete canonique :", arete)

# Verifier que (A, C) est reconnue comme la meme arete
meme_arete = tuple(sorted(("A", "C")))
print("Deja vue ?", meme_arete in aretes_vues)
```

## 8. Les fonctions

Une fonction est un bloc de code reutilisable. Elle prend des **parametres** en entree et retourne un **resultat**. Ecrire des fonctions rend le code modulaire, testable et lisible.

La norme PEP 8 (le guide de style Python) recommande :
- des noms de fonctions en `snake_case`
- une docstring pour expliquer ce que fait la fonction

```python
def calculer_degre(graphe, sommet):
    """
    Retourne le degre d'un sommet dans un graphe.

    Le degre est le nombre d'aretes incidentes au sommet.
    """
    return len(graphe[sommet])


graphe = {
    "A": [("B", 10), ("C", 5)],
    "B": [("A", 10)],
    "C": [("A", 5)],
}

print(f"Degre de A : {calculer_degre(graphe, 'A')}")
print(f"Degre de B : {calculer_degre(graphe, 'B')}")
```

```python
# Une fonction peut retourner plusieurs valeurs (tuple)
def statistiques_degres(graphe):
    """Retourne le degre minimal, maximal et moyen du graphe."""
    degres = [len(voisins) for voisins in graphe.values()]
    return min(degres), max(degres), sum(degres) / len(degres)


d_min, d_max, d_moy = statistiques_degres(graphe)
print(f"Degre min : {d_min}, max : {d_max}, moyen : {d_moy:.1f}")
```

```python
# Parametres avec valeur par defaut
def afficher_tournee(tournee, separateur=" -> "):
    """Affiche une tournee avec un separateur personnalisable."""
    print(separateur.join(tournee))


afficher_tournee(["Paris", "Lyon", "Marseille", "Paris"])
afficher_tournee(["A", "C", "B", "A"], separateur=" -- ")
```

## 9. Comprehensions de listes et de dictionnaires

Les comprehensions sont une syntaxe concise pour creer des listes ou des dictionnaires a partir d'une expression et d'une boucle. Elles sont tres utilisees en Python scientifique.

```python
# Comprehension de liste : les carres de 0 a 9
carres = [x ** 2 for x in range(10)]
print("Carres :", carres)

# Avec un filtre : garder uniquement les carres pairs
carres_pairs = [x ** 2 for x in range(10) if x % 2 == 0]
print("Carres pairs :", carres_pairs)
```

```python
# Application : extraire les noms des voisins d'un sommet
graphe = {
    "A": [("B", 10), ("C", 5), ("D", 8)],
    "B": [("A", 10)],
    "C": [("A", 5)],
    "D": [("A", 8)],
}

noms_voisins_a = [voisin for voisin, _distance in graphe["A"]]
print("Voisins de A :", noms_voisins_a)

# Trier les voisins de A par distance croissante
voisins_tries = sorted(graphe["A"], key=lambda x: x[1])
print("Voisins de A tries par distance :", voisins_tries)
```

```python
# Comprehension de dictionnaire
noms = ["A", "B", "C", "D"]
demandes = [12, 7, 19, 3]

# Associer chaque ville a sa demande
demandes_par_ville = {nom: demande for nom, demande in zip(noms, demandes)}
print(demandes_par_ville)

# Filtrer : garder seulement les villes avec une demande > 10
grosses_demandes = {k: v for k, v in demandes_par_ville.items() if v > 10}
print("Demandes > 10 :", grosses_demandes)
```

## 10. Les listes de listes : matrices

Pour representer une matrice d'adjacence ou une matrice de distances, on utilise une liste de listes (tableau 2D).

```python
# Creer une matrice 3x3 initialisee a 0
n = 3
matrice = [[0 for _ in range(n)] for _ in range(n)]

# Remplir : distance entre ville 0 et ville 1 = 15
matrice[0][1] = 15
matrice[1][0] = 15  # Symetrie pour un graphe non oriente

# Afficher la matrice ligne par ligne
for ligne in matrice:
    print(ligne)
```

### Piege classique : la copie par reference

Attention a la creation de matrices ! La multiplication d'une liste ne cree pas de copies independantes.

```python
# MAUVAISE facon : toutes les lignes partagent la meme reference
mauvaise_matrice = [[0] * 3] * 3
mauvaise_matrice[0][1] = 99
print("Mauvaise matrice (bug) :")
for ligne in mauvaise_matrice:
    print(ligne)

print()

# BONNE facon : chaque ligne est un objet independant
bonne_matrice = [[0] * 3 for _ in range(3)]
bonne_matrice[0][1] = 99
print("Bonne matrice :")
for ligne in bonne_matrice:
    print(ligne)
```

## 12. Gestion des erreurs

Un algorithme robuste verifie ses donnees d'entree. Python utilise les exceptions pour signaler les erreurs. Le bloc `try/except` permet de les intercepter.

```python
# Acceder a une cle inexistante dans un dictionnaire
graphe = {"A": [("B", 10)], "B": [("A", 10)]}

try:
    voisins = graphe["Z"]
except KeyError:
    print("Erreur : le sommet Z n'existe pas dans le graphe.")
```

```python
# Lever ses propres erreurs pour valider les entrees
def valider_graphe(graphe):
    """
    Verifie que chaque voisin reference existe et que les distances sont positives.
    """
    sommets = set(graphe.keys())

    for sommet, voisins in graphe.items():
        if not isinstance(voisins, list):
            raise TypeError(f"Les voisins de {sommet} doivent etre une liste.")

        for voisin, distance in voisins:
            if voisin not in sommets:
                raise ValueError(f"Le sommet {voisin} est reference mais non defini.")
            if distance < 0:
                raise ValueError(f"Distance negative ({distance}) entre {sommet} et {voisin}.")

    return True


graphe_ok = {
    "A": [("B", 10), ("C", 5)],
    "B": [("A", 10)],
    "C": [("A", 5)],
}

print("Graphe valide :", valider_graphe(graphe_ok))

# Test avec un graphe invalide
graphe_invalide = {"A": [("X", 10)]}

try:
    valider_graphe(graphe_invalide)
except ValueError as e:
    print(f"Erreur detectee : {e}")
```

## 13. Synthese : construire un petit reseau routier

Cet exercice combine toutes les notions vues precedemment pour construire et analyser un reseau de villes avec des coordonnees aleatoires.

```python
import math
import random

random.seed(42)

# 1. Generer 5 villes avec des coordonnees aleatoires
noms = ["Paris", "Lyon", "Marseille", "Nantes", "Lille"]
coordonnees = {nom: (random.randint(0, 100), random.randint(0, 100)) for nom in noms}

print("Coordonnees des villes :")
for nom, (x, y) in coordonnees.items():
    print(f"  {nom:>10} : ({x:>3}, {y:>3})")
```

```python
# 2. Calculer toutes les distances entre paires de villes
def distance_euclidienne(point_a, point_b):
    xa, ya = point_a
    xb, yb = point_b
    return math.sqrt((xb - xa) ** 2 + (yb - ya) ** 2)


distances = {}
for ville_a in noms:
    for ville_b in noms:
        if ville_a != ville_b:
            d = round(distance_euclidienne(coordonnees[ville_a], coordonnees[ville_b]), 2)
            distances[(ville_a, ville_b)] = d

print("\nQuelques distances :")
paires_affichees = 0
for (va, vb), d in distances.items():
    if va < vb:  # Eviter d'afficher A-B et B-A
        print(f"  {va:>10} -- {vb:<10} : {d:>7.2f}")
        paires_affichees += 1

print(f"\nNombre total de paires : {paires_affichees}")
```

```python
# 3. Construire le graphe complet (liste d'adjacence)
graphe_complet = {}
for ville in noms:
    graphe_complet[ville] = [
        (autre, distances[(ville, autre)])
        for autre in noms
        if autre != ville
    ]

# 4. Verifier quelques proprietes
print("Proprietes du graphe :")
print(f"  Nombre de sommets : {len(graphe_complet)}")

nb_aretes = sum(len(v) for v in graphe_complet.values()) // 2
print(f"  Nombre d'aretes : {nb_aretes}")

for sommet, voisins in graphe_complet.items():
    plus_proche = min(voisins, key=lambda x: x[1])
    print(f"  Plus proche voisin de {sommet:>10} : {plus_proche[0]} ({plus_proche[1]} km)")
```

```python
# 5. Construire la matrice d'adjacence equivalente
n = len(noms)
indice = {nom: i for i, nom in enumerate(noms)}

matrice_adj = [[0.0] * n for _ in range(n)]
for (va, vb), d in distances.items():
    matrice_adj[indice[va]][indice[vb]] = d

# Afficher la matrice
print(f"{'':>12}", end="")
for nom in noms:
    print(f"{nom:>12}", end="")
print()

for i, nom in enumerate(noms):
    print(f"{nom:>12}", end="")
    for j in range(n):
        print(f"{matrice_adj[i][j]:>12.2f}", end="")
    print()
```
