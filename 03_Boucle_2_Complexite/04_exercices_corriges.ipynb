# Exercices sur la complexité algorithmique

## Niveau 1 -- Compréhension intuitive

### Exercice 1 -- Comptage de boucles simples

```python
for i in range(n):
    print(i)
```

**Question** : Quelle est la complexité en temps de ce code ?

````python
# Expérimentez : mesurez le temps d'exécution pour différentes valeurs de n
import time

def exercice_1(n):
    for i in range(n):
        _ = i  # opération O(1)

# Testez avec différentes valeurs de n
for n in [1000, 10000, 100000, 1000000]:
    debut = time.time()
    exercice_1(n)
    duree = time.time() - debut
    print(f"n = {n:>10} -> temps = {duree:.6f} s")
````

**Correction** :

- La boucle s'exécute `n` fois.
- L'instruction dans la boucle est `O(1)` -> coût constant.
- Donc **complexité** : `O(n)`

### Exercice 2 -- Boucle imbriquée

```python
for i in range(n):
    for j in range(n):
        print(i, j)
```

**Question** : Quelle est la complexité temporelle ?

````python
# Expérimentez : observez la croissance quadratique
import time

def exercice_2(n):
    compteur = 0
    for i in range(n):
        for j in range(n):
            compteur += 1
    return compteur

for n in [100, 200, 500, 1000]:
    debut = time.time()
    operations = exercice_2(n)
    duree = time.time() - debut
    print(f"n = {n:>5} -> opérations = {operations:>10} (n² = {n*n:>10}) -> temps = {duree:.6f} s")
````

**Correction** :

- La boucle intérieure s'exécute `n` fois pour **chaque** itération de la boucle extérieure (qui s'exécute aussi `n` fois).
- Donc total : `n x n = n²`
- **Complexité** : `O(n²)`

### Exercice 3 -- Logarithmique

```python
i = 1
while i < n:
    print(i)
    i *= 2
```

**Question** : Quelle est la complexité en temps ?

````python
# Expérimentez : comptez les itérations et comparez avec log2(n)
import math

def exercice_3(n):
    i = 1
    iterations = 0
    while i < n:
        i *= 2
        iterations += 1
    return iterations

for n in [10, 100, 1000, 10000, 100000, 1000000]:
    iterations = exercice_3(n)
    log_n = math.log2(n)
    print(f"n = {n:>10} -> itérations = {iterations:>3} -> log2(n) = {log_n:.1f}")
````

**Correction** :

- À chaque itération, `i` double -> `i = 1, 2, 4, 8, ..., <= n`
- Il faut environ `log2(n)` étapes avant que `i >= n`
- **Complexité** : `O(log n)`

## Niveau 2 -- Analyse de cas intermédiaires

### Exercice 4 -- Complexité mixte

```python
for i in range(n):
    for j in range(1, n, 2):
        print(i, j)
```

**Question** : Quelle est la complexité ?

````python
# Votre espace d'expérimentation
# Comptez les opérations pour vérifier votre analyse

def exercice_4(n):
    compteur = 0
    for i in range(n):
        for j in range(1, n, 2):
            compteur += 1
    return compteur

for n in [100, 200, 500, 1000]:
    ops = exercice_4(n)
    print(f"n = {n:>5} -> opérations = {ops:>10} -> n*n/2 = {n*n//2:>10}")
````

**Correction** :

- La boucle extérieure : `n` itérations
- La boucle intérieure saute de 2 en 2 -> environ `n/2` itérations
- Total environ `n x (n/2) = O(n²)`

### Exercice 5 -- Cas avec dépendance

```python
for i in range(n):
    for j in range(i):
        print(i, j)
```

**Question** : Quelle est la complexité ?

````python
# Votre espace d'expérimentation
# Observez la somme 0 + 1 + 2 + ... + (n-1)

def exercice_5(n):
    compteur = 0
    for i in range(n):
        for j in range(i):
            compteur += 1
    return compteur

for n in [10, 50, 100, 500, 1000]:
    ops = exercice_5(n)
    formule = n * (n - 1) // 2
    print(f"n = {n:>5} -> opérations = {ops:>10} -> n(n-1)/2 = {formule:>10}")
````

**Correction** :

- Pour `i = 0` -> 0 itérations
- Pour `i = 1` -> 1
- ...
- Total = `0 + 1 + 2 + ... + (n-1)` = `n(n-1)/2`
- Donc **complexité** : `O(n²)`

## Niveau 3 -- Fonctions récursives

### Exercice 6 -- Récursion simple

```python
def f(n):
    if n == 0:
        return
    f(n - 1)
```

**Question** : Combien d'appels récursifs sont faits ? Complexité ?

````python
# Expérimentez : comptez les appels

def f_exercice_6(n, compteur):
    compteur[0] += 1
    if n == 0:
        return
    f_exercice_6(n - 1, compteur)

for n in [5, 10, 50, 100]:
    compteur = [0]
    f_exercice_6(n, compteur)
    print(f"n = {n:>5} -> appels = {compteur[0]}")
````

**Correction** :

- La fonction s'appelle elle-même `n` fois avant d'atteindre le cas de base.
- Chaque appel prend `O(1)`
- **Complexité** : `O(n)`

### Exercice 7 -- Récursion double (exponentielle)

```python
def f(n):
    if n <= 1:
        return 1
    return f(n - 1) + f(n - 2)
```

**Question** : Quelle est la complexité ?

````python
# Expérimentez : observez la croissance exponentielle des appels

def fib_naive(n, compteur):
    compteur[0] += 1
    if n <= 1:
        return 1
    return fib_naive(n - 1, compteur) + fib_naive(n - 2, compteur)

print("Fibonacci naïf - nombre d'appels :")
for n in [5, 10, 15, 20, 25, 30]:
    compteur = [0]
    fib_naive(n, compteur)
    print(f"n = {n:>3} -> appels = {compteur[0]:>12} -> ~2^n = {2**n:>12}")
````

**Correction** :

- C'est la définition naïve de Fibonacci.
- L'arbre d'appel contient environ `2^n` appels -> croissance exponentielle.
- **Complexité** : `O(2^n)`

### Exercice 8 -- Récursion logarithmique

```python
def f(n):
    if n == 0:
        return
    f(n // 2)
```

**Question** : Combien d'appels récursifs sont effectués ?

````python
# Expérimentez

import math

def f_exercice_8(n, compteur):
    compteur[0] += 1
    if n == 0:
        return
    f_exercice_8(n // 2, compteur)

for n in [10, 100, 1000, 10000, 1000000]:
    compteur = [0]
    f_exercice_8(n, compteur)
    log_val = math.log2(n) if n > 0 else 0
    print(f"n = {n:>10} -> appels = {compteur[0]:>3} -> log2(n) = {log_val:.1f}")
````

**Correction** :

- À chaque appel, `n` est divisé par 2
- Nombre d'appels environ `log2(n)`
- **Complexité** : `O(log n)`

## Niveau 4 -- Bonus de réflexion

### Exercice 9 -- Analyse amortie

```python
arr = []
for i in range(n):
    arr.append(i)
```

**Question** : En Python, `append()` peut parfois causer une réallocation mémoire. Quelle est la complexité **amortie** de cet algorithme ?

````python
# Expérimentez : observez le temps d'exécution de append()
import time
import sys

def exercice_9(n):
    arr = []
    tailles_observees = []
    for i in range(n):
        taille_avant = sys.getsizeof(arr)
        arr.append(i)
        taille_apres = sys.getsizeof(arr)
        if taille_apres > taille_avant:
            tailles_observees.append((i, taille_avant, taille_apres))
    return tailles_observees

reallocs = exercice_9(100)
print("Réallocations mémoire observées :")
print(f"{'Index':>8} {'Taille avant':>15} {'Taille après':>15}")
for idx, avant, apres in reallocs:
    print(f"{idx:>8} {avant:>15} {apres:>15}")

print(f"\nNombre de réallocations pour 100 appels append() : {len(reallocs)}")
print("=> La plupart des appels sont O(1), les réallocations sont rares.")
````

**Correction** :

- `append()` est généralement `O(1)` **amorti** (même si la réallocation coûte `O(n)` ponctuellement).
- Donc sur `n` appels : **O(n)** en **temps amorti**

### Exercice 10 -- Code mélangeant plusieurs types de complexité

```python
for i in range(n):
    j = 1
    while j < n:
        j *= 2
        print(i, j)
```

**Question** : Complexité en temps ?

````python
# Expérimentez : vérifiez la complexité O(n log n)
import math

def exercice_10(n):
    compteur = 0
    for i in range(n):
        j = 1
        while j < n:
            j *= 2
            compteur += 1
    return compteur

print(f"{'n':>8} {'opérations':>12} {'n*log2(n)':>12} {'ratio':>8}")
print("-" * 44)
for n in [100, 500, 1000, 5000, 10000]:
    ops = exercice_10(n)
    n_log_n = n * math.log2(n)
    ratio = ops / n_log_n if n_log_n > 0 else 0
    print(f"{n:>8} {ops:>12} {n_log_n:>12.0f} {ratio:>8.2f}")
````

**Correction** :

- La boucle `while` est `O(log n)` (j double à chaque fois)
- Elle est répétée `n` fois (dans la boucle `for`)
- Donc **complexité totale** : `O(n log n)`
