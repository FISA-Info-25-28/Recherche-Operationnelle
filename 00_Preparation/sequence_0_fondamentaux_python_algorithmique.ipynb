## 1. Les collections : listes et dictionnaires

Pour representer un reseau routier ou une ville, il est necessaire de stocker l'information en memoire de maniere optimisee. Python propose nativement des structures adaptees.

**Les listes de listes (tableaux a deux dimensions)**
Elles sont particulierement utiles pour creer une matrice d'adjacence, c'est-a-dire un tableau croise indiquant la presence et le cout d'une arete entre deux sommets.

**Les dictionnaires**
Les dictionnaires permettent d'associer une cle a une valeur. Ils sont ideaux pour implementer une liste d'adjacence, la methode la plus repandue pour stocker un graphe en memoire de maniere compacte.

```python
# Creation d'une matrice 3x3 initialisee a 0
nombre_villes = 3
matrice_distances = [[0 for _ in range(nombre_villes)] for _ in range(nombre_villes)]

# Assignation d'une distance de 15 entre la ville 0 et la ville 1
matrice_distances[0][1] = 15
matrice_distances[1][0] = 15  # Symetrie pour un trajet a double sens

print(matrice_distances)

# Representation d'un reseau sous forme de dictionnaire
# La cle est le sommet, la valeur est une liste de tuples (sommet_voisin, distance)
graphe_routier = {
    "A": [("B", 10), ("C", 5)],
    "B": [("A", 10)],
    "C": [("A", 5)]
}

# Acceder aux voisins de la ville A
voisins_a = graphe_routier["A"]
print(voisins_a)
```

### Point d'attention : copie de structures

Lorsqu'on manipule des matrices, une erreur frequente consiste a dupliquer plusieurs fois la meme ligne en memoire. Il faut donc privilegier les comprehensions de listes plutot qu'une multiplication naive.

```python
taille = 3
mauvaise_matrice = [[0] * taille] * taille
bonne_matrice = [[0] * taille for _ in range(taille)]

mauvaise_matrice[0][1] = 99
bonne_matrice[0][1] = 99

print("Mauvaise matrice :", mauvaise_matrice)
print("Bonne matrice :", bonne_matrice)
```

## 2. Structures de controle : iterations et conditions

Les algorithmes d'optimisation necessitent de parcourir de grandes quantites de donnees et d'effectuer des tests logiques, par exemple verifier si le degre d'un sommet est pair ou impair.

```python
# Exemple : calculer le degre de chaque sommet dans une liste d'adjacence
for sommet, voisins in graphe_routier.items():
    degre = len(voisins)

    if degre % 2 == 0:
        print(f"Le sommet {sommet} a un degre pair ({degre}).")
    else:
        print(f"Le sommet {sommet} a un degre impair ({degre}).")
```

## 3. Fonctions et modularite (norme PEP 8)

Il est attendu que le code produit soit lisible, modulaire et commente, en respectant les standards de la norme PEP 8. L'utilisation de fonctions avec des chaines de documentation est importante pour structurer les algorithmes.

```python
def calculer_degre_maximal(graphe):
    """
    Parcourt un graphe represente par un dictionnaire et retourne le degre maximal.

    Parametres
    ----------
    graphe : dict
        Le graphe sous forme de liste d'adjacence.

    Retourne
    --------
    int
        La valeur du degre le plus eleve.
    """
    if not graphe:
        return 0

    return max(len(voisins) for voisins in graphe.values())


print(calculer_degre_maximal(graphe_routier))
```

**Mesure du temps d'execution**
Afin de determiner la performance temporelle des algorithmes et tracer leurs courbes de complexite, il est necessaire de mesurer le temps processeur alloue a l'execution du programme.

```python
import math
import random
import time

random.seed(42)

# Generation d'entiers aleatoires entre 0 et 100
coordonnee_x = random.randint(0, 100)
coordonnee_y = random.randint(0, 100)
print("Coordonnees generees :", coordonnee_x, coordonnee_y)

# Calcul d'une distance euclidienne
distance = math.sqrt((10 - 5) ** 2 + (20 - 10) ** 2)
print("Distance calculee :", round(distance, 2))

# Mesure d'un temps CPU
debut_cpu = time.process_time()
somme = 0
for i in range(100000):
    somme += i
fin_cpu = time.process_time()

temps_execution = (fin_cpu - debut_cpu) * 1000
print("Somme calculee :", somme)
print(f"Temps CPU : {temps_execution:.3f} ms")
```

## 5. Ensembles et tuples : memoriser sans doublons

Dans de nombreux algorithmes sur les graphes, il faut memoriser les sommets deja visites ou garantir qu'une arete n'est pas comptee plusieurs fois. Les ensembles (`set`) sont adaptes a cet usage. Les tuples sont quant a eux utiles pour representer des couples de valeurs de maniere compacte.

```python
sommets_visites = set()
sommets_visites.add("A")
sommets_visites.add("B")
sommets_visites.add("A")  # Aucun doublon ajoute
print(sommets_visites)

# Representation canonique d'une arete non orientee
arete = tuple(sorted(("C", "A")))
print(arete)
```

## 6. Comprehensions, tri et fonctions utiles

Les comprehensions permettent d'ecrire des transformations de donnees de maniere concise. Des fonctions comme `sorted`, `min`, `max`, `enumerate` ou `zip` sont tres utiles pour manipuler des structures de donnees en algorithmique.

```python
# Extraire uniquement les noms des voisins de A
noms_voisins = [voisin for voisin, _distance in graphe_routier["A"]]
print(noms_voisins)

# Trier les voisins par distance croissante
voisins_tries = sorted(graphe_routier["A"], key=lambda element: element[1])
print(voisins_tries)

# Associer un indice a chaque sommet
for indice, sommet in enumerate(graphe_routier.keys()):
    print(indice, sommet)

# Construire un dictionnaire a partir de deux listes
noms = ["A", "B", "C"]
demandes = [12, 7, 19]
demandes_par_ville = dict(zip(noms, demandes))
print(demandes_par_ville)
```

## 7. Validation des donnees et robustesse

Un algorithme fiable commence par verifier que les donnees d'entree respectent certaines hypotheses. Dans le cadre d'un graphe, on peut par exemple verifier que chaque distance est positive et que chaque sommet mentionne existe bien.

```python
def valider_graphe(graphe):
    """
    Verifie quelques proprietes simples d'un graphe pondere.

    Parametres
    ----------
    graphe : dict
        Le graphe sous forme de liste d'adjacence.

    Retourne
    --------
    bool
        True si la validation reussit.

    Raises
    ------
    TypeError
        Si les voisins d'un sommet ne sont pas stockes dans une liste.
    ValueError
        Si un sommet reference n'existe pas ou si une distance est negative.
    """
    sommets = set(graphe.keys())

    for sommet, voisins in graphe.items():
        if not isinstance(voisins, list):
            raise TypeError(f"Les voisins de {sommet} doivent etre stockes dans une liste.")

        for voisin, distance in voisins:
            if voisin not in sommets:
                raise ValueError(f"Le sommet {voisin} est reference mais non defini.")
            if distance < 0:
                raise ValueError("Une distance ne peut pas etre negative.")

    return True


print(valider_graphe(graphe_routier))
```

## 8. Exemple de synthese : manipuler des coordonnees

Pour preparer les boucles suivantes, on peut combiner plusieurs notions deja vues : dictionnaires, fonctions, comprehensions et calcul de distance. Cet exemple construit un petit ensemble de villes, puis calcule les distances entre certaines paires.

```python
import math

villes = {
    "A": (0, 0),
    "B": (3, 4),
    "C": (6, 8)
}


def distance_euclidienne(point_a, point_b):
    """Calcule la distance euclidienne entre deux points du plan."""
    x_a, y_a = point_a
    x_b, y_b = point_b
    return math.sqrt((x_b - x_a) ** 2 + (y_b - y_a) ** 2)


distances = {
    (ville_depart, ville_arrivee): distance_euclidienne(coords_depart, coords_arrivee)
    for ville_depart, coords_depart in villes.items()
    for ville_arrivee, coords_arrivee in villes.items()
    if ville_depart != ville_arrivee
}

for trajet, distance in distances.items():
    print(trajet, round(distance, 2))
```
