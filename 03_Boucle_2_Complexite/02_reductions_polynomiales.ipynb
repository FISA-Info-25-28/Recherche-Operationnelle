# Réduction polynomiale

## Pourquoi parle-t-on de réduction polynomiale ?

### Définition intuitive

Lorsqu'on fait face à un problème difficile à résoudre directement, une approche classique consiste à le **transformer** en un autre problème pour lequel on connaît une méthode de résolution efficace. C'est le principe des **réductions polynomiales**.

Une **réduction polynomiale** d'un problème A à un problème B consiste à transformer **chaque instance du problème A** en une **instance équivalente du problème B**, de manière à ce que les réponses soient les mêmes (oui ou non), et que cette transformation soit **effectuée en temps polynomial** par rapport à la taille de l'entrée.

Autrement dit, plutôt que de chercher une solution directe à A, on "traduira" A en B, résoudra B, puis interprètera la solution pour répondre à A.

Cela permet notamment d'**évaluer la difficulté relative** entre deux problèmes. Si on parvient à réduire A à B efficacement, alors **résoudre B permet aussi de résoudre A**.

### Définition formelle

On dit qu'un problème A se **réduit en temps polynomial** à un problème B, ce que l'on note **A ≤p B**, s'il existe une **fonction de transformation f**, calculable en temps polynomial, telle que :

- Pour toute instance x de A, f(x) est une instance du problème B ;
- x est une instance dont la réponse est "oui" pour A **si et seulement si** f(x) est une instance dont la réponse est "oui" pour B.

Cette définition implique deux conditions essentielles :

1. **Temps polynomial** : la transformation de x en f(x) doit pouvoir se faire efficacement, c'est-à-dire en un nombre d'étapes borné par un polynôme en la taille de x. Cela garantit que la réduction elle-même ne crée pas un nouveau problème difficile à calculer.

2. **Conservation de la réponse** : il ne suffit pas de transformer le problème ; il faut aussi que la solution soit conservée. La réponse obtenue pour B doit correspondre à la réponse pour A.

Grâce à cette construction, si l'on dispose d'un algorithme pour résoudre B (même si B est un problème complètement différent de A en apparence), on peut **résoudre A en combinant cette transformation et la résolution de B**.

### Intuition et intérêt pratique

Ce concept est fondamental en algorithmique, et notamment dans l'étude des problèmes de complexité **NP** et **NP-complets**. Voici pourquoi :

- Une réduction polynomiale permet de montrer que **résoudre un problème A n'est pas plus dur que résoudre un autre problème B**.
- En particulier, si A est un problème pour lequel **on ne connaît pas d'algorithme efficace**, et qu'on parvient à le réduire à un autre problème B qui **est bien connu et supposé difficile**, cela donne **une indication forte sur la complexité de A**.
- C'est ainsi qu'ont été identifiés les **problèmes NP-complets** : ceux vers lesquels on peut réduire en temps polynomial tous les problèmes de NP.

### Notation et vocabulaire

- **A** et **B** désignent des **problèmes de décision** (dont la réponse est "oui" ou "non"), comme par exemple le TSP ou le problème du sac à dos.
- L'écriture **A ≤p B** signifie que **le problème A se transforme en B** en **temps polynomial**.
- Une **réduction polynomiale** garantit que la transformation d'une instance de A en une instance de B se fait en un **temps raisonnable** (borné par un polynôme en fonction de la taille de l'entrée).
- Une réduction est dite **correcte** si l'instance transformée donne **la même réponse** que l'originale : si A répond "oui", alors B aussi.
- Une réduction est dite **complète** si elle peut être appliquée à **tout problème d'une classe de complexité donnée** (comme tous les problèmes de NP).

## 2. Intérêt des réductions

Les réductions polynomiales jouent un rôle fondamental en algorithmique et en théorie de la complexité. Elles permettent de relier les problèmes entre eux, de les comparer en termes de difficulté, et de mieux comprendre les limites des algorithmes que nous pouvons concevoir.

Voici les principaux intérêts :

- **Identifier des problèmes NP-complets** : Lorsqu'on arrive à réduire un problème connu pour être difficile (déjà prouvé NP-complet) à un nouveau problème, cela permet de montrer que ce dernier est **au moins aussi difficile**. Si le nouveau problème est aussi dans la classe NP, on peut alors conclure qu'il est **NP-complet**.

- **Prouver qu'un problème n'a probablement pas d'algorithme efficace** : Si un problème peut être réduit à un autre problème NP-complet, cela suggère qu'il est **peu probable qu'il puisse être résolu en temps polynomial**, à moins que P = NP.

- **Trouver des stratégies de résolution indirectes** : Parfois, au lieu d'inventer un nouvel algorithme pour un problème donné, il est plus judicieux de le reformuler (par une réduction) en un problème pour lequel il existe déjà des outils efficaces ou bien étudiés.

- **Utiliser les réductions dans des raisonnements par l'absurde** : Les réductions sont souvent utilisées dans des **preuves par contradiction**. Si on suppose qu'un problème A peut être résolu efficacement, alors, par réduction, on pourrait aussi résoudre un problème B que l'on sait être difficile.

## 3. Exemple simple : réduction du problème de la clique au problème de la couverture de sommets

### Description des deux problèmes

- **Problème de la clique** : étant donné un graphe non orienté G = (V, E) et un entier k, existe-t-il un sous-ensemble de k sommets dans G formant une **clique** ?
  Autrement dit, peut-on trouver k sommets tels que **chaque paire de sommets de cet ensemble soit reliée par une arête** dans G ?

- **Problème de la couverture de sommets** (*Vertex Cover*) : étant donné un graphe G = (V, E) et un entier k, existe-t-il un ensemble de k sommets qui **touche toutes les arêtes du graphe** ?
  Autrement dit, peut-on choisir k sommets tels que **chaque arête du graphe ait au moins un de ses deux sommets dans cet ensemble** ?

### Objectif et principe de la réduction

Nous allons montrer que si nous savons résoudre efficacement le problème de la couverture de sommets, alors nous pouvons aussi résoudre efficacement le problème de la clique, grâce à une réduction polynomiale.

Voici les grandes étapes de la transformation :

1. On part d'un graphe G = (V, E) et d'un entier k (la taille de la clique recherchée).
2. On construit le **complémentaire** de G, noté G', c'est-à-dire un graphe qui a les **mêmes sommets**, mais où une arête est présente dans G' si et seulement si elle **n'est pas présente dans G**.
3. On cherche alors une **couverture de sommets** de taille **|V| - k** dans G'.

### Justification

- Si G contient une clique de taille k, alors il existe un ensemble de |V| - k sommets qui couvrent toutes les arêtes de G'.
- Inversement, si G' possède une couverture de sommets de taille |V| - k, alors les k sommets restants forment une clique dans G.

```python
# Implémentation de la réduction Clique -> Vertex Cover

def construire_complement(sommets, aretes):
    """Construit le graphe complémentaire."""
    toutes_aretes = set()
    for i in range(len(sommets)):
        for j in range(i + 1, len(sommets)):
            toutes_aretes.add((sommets[i], sommets[j]))
    # Normaliser les arêtes d'entrée
    aretes_norm = set()
    for u, v in aretes:
        aretes_norm.add((min(u, v), max(u, v)))
    complement = toutes_aretes - aretes_norm
    return complement

def est_clique(sommets_candidats, aretes):
    """Vérifie si les sommets forment une clique."""
    aretes_norm = set()
    for u, v in aretes:
        aretes_norm.add((min(u, v), max(u, v)))
    for i in range(len(sommets_candidats)):
        for j in range(i + 1, len(sommets_candidats)):
            u, v = min(sommets_candidats[i], sommets_candidats[j]), max(sommets_candidats[i], sommets_candidats[j])
            if (u, v) not in aretes_norm:
                return False
    return True

def est_vertex_cover(sommets_candidats, aretes):
    """Vérifie si les sommets forment une couverture de sommets."""
    cover = set(sommets_candidats)
    for u, v in aretes:
        if u not in cover and v not in cover:
            return False
    return True

# Exemple
sommets = [1, 2, 3, 4, 5]
aretes_G = [(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)]

print("=== Graphe G ===")
print(f"Sommets : {sommets}")
print(f"Arêtes  : {aretes_G}")

# Clique {1, 2, 3} dans G
clique = [1, 2, 3]
k = len(clique)
print(f"\nClique candidate : {clique}")
print(f"Est-ce une clique ? {est_clique(clique, aretes_G)}")

# Réduction : construire le complémentaire
aretes_complement = construire_complement(sommets, aretes_G)
print(f"\n=== Graphe complémentaire G' ===")
print(f"Arêtes de G' : {aretes_complement}")

# Vertex cover de taille |V| - k dans G'
vertex_cover = [s for s in sommets if s not in clique]
print(f"\nVertex Cover dans G' (sommets hors clique) : {vertex_cover}")
print(f"Taille attendue : |V| - k = {len(sommets)} - {k} = {len(sommets) - k}")
print(f"Est-ce un vertex cover de G' ? {est_vertex_cover(vertex_cover, aretes_complement)}")
```

## 4. Exemple : réduction du problème du voyageur de commerce (TSP) au problème du cycle hamiltonien

### Présentation des deux problèmes

- **Problème du voyageur de commerce (TSP)** : on dispose d'un graphe complet pondéré (chaque arête a un coût). Le but est de trouver un cycle qui passe une seule fois par chaque sommet, revient au point de départ, et dont le **coût total est inférieur ou égal à une borne k**.

- **Problème du cycle hamiltonien** : étant donné un graphe non pondéré, peut-on trouver un cycle simple qui **passe une et une seule fois par chaque sommet**, et revient au point de départ ?

### Construction de la réduction

1. On part d'un graphe non pondéré G = (V, E), instance du problème du cycle hamiltonien.
2. On construit un graphe complet G' = (V, E'), contenant **toutes les arêtes possibles** entre les sommets.
3. On assigne un **poids aux arêtes** :
   - Arête existant dans G : poids **1**
   - Arête ajoutée (absente de G) : poids **2**
4. On fixe la borne k = n (nombre de sommets).

### Justification

- Si G possède un cycle hamiltonien, ce cycle n'utilise que des arêtes de poids 1 dans G', donc coût total = n.
- Si G ne possède pas de cycle hamiltonien, toute tournée dans G' utilisera au moins une arête de poids 2, donc coût > n.

```python
# Illustration de la réduction Cycle Hamiltonien -> TSP
from itertools import permutations

def reduction_hamiltonien_vers_tsp(sommets, aretes):
    """Transforme un graphe non pondéré en une instance TSP."""
    # Normaliser les arêtes
    aretes_set = set()
    for u, v in aretes:
        aretes_set.add((min(u, v), max(u, v)))
    
    # Construire le graphe complet pondéré
    poids = {}
    for i in range(len(sommets)):
        for j in range(i + 1, len(sommets)):
            u, v = sommets[i], sommets[j]
            arete = (min(u, v), max(u, v))
            poids[arete] = 1 if arete in aretes_set else 2
    
    borne_k = len(sommets)
    return poids, borne_k

def tsp_brute_force(sommets, poids):
    """Résout le TSP par force brute."""
    meilleur_cout = float('inf')
    meilleur_tour = None
    depart = sommets[0]
    
    for perm in permutations(sommets[1:]):
        tour = [depart] + list(perm)
        cout = 0
        for i in range(len(tour)):
            u = tour[i]
            v = tour[(i + 1) % len(tour)]
            arete = (min(u, v), max(u, v))
            cout += poids[arete]
        if cout < meilleur_cout:
            meilleur_cout = cout
            meilleur_tour = tour
    
    return meilleur_tour, meilleur_cout

# Exemple 1 : graphe AVEC cycle hamiltonien
print("=== Graphe avec cycle hamiltonien ===")
sommets = [1, 2, 3, 4]
aretes = [(1, 2), (2, 3), (3, 4), (4, 1)]  # Cycle 1-2-3-4-1
print(f"Sommets : {sommets}")
print(f"Arêtes  : {aretes}")

poids, borne = reduction_hamiltonien_vers_tsp(sommets, aretes)
print(f"Poids après réduction : {poids}")
print(f"Borne k = {borne}")

tour, cout = tsp_brute_force(sommets, poids)
print(f"Meilleur tour TSP : {tour} -> {tour[0]}, coût = {cout}")
print(f"Cycle hamiltonien existe ? {cout <= borne}")

# Exemple 2 : graphe SANS cycle hamiltonien
print("\n=== Graphe sans cycle hamiltonien ===")
sommets2 = [1, 2, 3, 4]
aretes2 = [(1, 2), (2, 3)]  # Pas assez d'arêtes
print(f"Sommets : {sommets2}")
print(f"Arêtes  : {aretes2}")

poids2, borne2 = reduction_hamiltonien_vers_tsp(sommets2, aretes2)
print(f"Poids après réduction : {poids2}")
print(f"Borne k = {borne2}")

tour2, cout2 = tsp_brute_force(sommets2, poids2)
print(f"Meilleur tour TSP : {tour2} -> {tour2[0]}, coût = {cout2}")
print(f"Cycle hamiltonien existe ? {cout2 <= borne2}")
```

## 5. Exemple : réduction du problème de la somme des sous-ensembles (Subset Sum) au problème du sac à dos (Knapsack)

### Présentation des deux problèmes

- **Problème de la somme des sous-ensembles (Subset Sum)** : étant donné un ensemble d'entiers positifs et une valeur cible T, existe-t-il un **sous-ensemble de ces entiers dont la somme est exactement égale à T** ?

- **Problème du sac à dos (Knapsack)** : on dispose d'un ensemble d'objets, chacun avec un **poids** et une **valeur**. Le problème est de savoir s'il existe un sous-ensemble d'objets **dont la somme des poids n'excède pas W** et **dont la somme des valeurs est au moins V**.

### Construction de la réduction

Soit une instance de Subset Sum donnée par un ensemble d'entiers {a1, a2, ..., an} et une valeur cible T.

Pour obtenir une instance équivalente de Knapsack :

1. Chaque entier ai représente à la fois le **poids** et la **valeur** d'un objet : wi = ai, vi = ai
2. Capacité maximale du sac : W = T
3. Valeur minimale souhaitée : V = T

Puisque poids = valeur pour tous les objets, Knapsack demande : existe-t-il un sous-ensemble dont la somme est exactement T ?

### Justification

- Subset Sum "oui" => sous-ensemble de somme T => poids total <= T et valeur >= T => Knapsack "oui"
- Knapsack "oui" => poids <= T et valeur >= T, mais poids = valeur => somme = T => Subset Sum "oui"

```python
# Implémentation de la réduction Subset Sum -> Knapsack

def subset_sum_vers_knapsack(entiers, cible):
    """Transforme une instance de Subset Sum en instance de Knapsack."""
    # Chaque entier devient un objet avec poids = valeur = entier
    objets = [(a, a) for a in entiers]  # (poids, valeur)
    W = cible  # Capacité
    V = cible  # Valeur minimale
    return objets, W, V

def resoudre_knapsack_decision(objets, W, V):
    """Résout le problème du sac à dos (décision) par force brute."""
    n = len(objets)
    for masque in range(1, 2**n):
        poids_total = 0
        valeur_totale = 0
        selection = []
        for i in range(n):
            if masque & (1 << i):
                poids_total += objets[i][0]
                valeur_totale += objets[i][1]
                selection.append(objets[i][0])
        if poids_total <= W and valeur_totale >= V:
            return True, selection
    return False, []

# Exemple
entiers = [3, 7, 1, 8, 4]
cible = 12

print(f"=== Subset Sum ===")
print(f"Entiers : {entiers}")
print(f"Cible T : {cible}")

# Réduction
objets, W, V = subset_sum_vers_knapsack(entiers, cible)
print(f"\n=== Instance Knapsack (après réduction) ===")
print(f"Objets (poids, valeur) : {objets}")
print(f"Capacité W = {W}, Valeur min V = {V}")

# Résolution
existe, selection = resoudre_knapsack_decision(objets, W, V)
print(f"\n=== Résultat ===")
print(f"Solution Knapsack : {existe}")
if existe:
    print(f"Sous-ensemble : {selection} (somme = {sum(selection)})")
    print(f"Donc Subset Sum a aussi une solution !")
```

```python
# Expérimentez avec vos propres valeurs !
# Modifiez les entiers et la cible ci-dessous

mes_entiers = [5, 10, 15, 20, 25]
ma_cible = 30

objets, W, V = subset_sum_vers_knapsack(mes_entiers, ma_cible)
existe, selection = resoudre_knapsack_decision(objets, W, V)

print(f"Entiers : {mes_entiers}, Cible : {ma_cible}")
print(f"Existe un sous-ensemble de somme {ma_cible} ? {existe}")
if existe:
    print(f"Sous-ensemble trouvé : {selection}")
```
